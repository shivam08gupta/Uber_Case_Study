# Converting date-time format#

```python
#for column 'Request timestamp'
uber['Request timestamp'] = pd.to_datetime(uber['Request timestamp'],errors='ignore', dayfirst=True)
uber
```

```python
#for column 'Drop timestamp'
uber['Drop timestamp'] = pd.to_datetime(uber['Drop timestamp'], errors='ignore', dayfirst=True)
uber
```

```python
#Extracting new columns from Request_timestamp and Drop_timestamp
uber['Pickup_time'] = uber['Request timestamp'].dt.time
uber['Drop_time'] = uber['Drop timestamp'].dt.time
uber['Request_day'] = uber['Request timestamp'].dt.day_name()
uber.head()
```

```python
#New Column- Request hour
uber['Request_hour'] = uber['Request timestamp'].dt.hour
uber.head()
```

```python
#Extracting Parts of the day
uber['Day_part'] = pd.cut(uber['Request_hour'],[0,4,8,12,16,20,24],labels=['Late Night', 'Early Morning', 'Morning', 'Afternoon', 'Evening', 'Night'])
uber.head()
```

```python
#Distributing Uber dataframe into 2dataframes i.e city and airport from column'Pickup point'
uber_city = uber.loc[uber['Pickup point'] == 'City']
uber_airport = uber.loc[uber['Pickup point'] == 'Airport']
```

# UBER STATUS FROM CITY TO AIRPORT (DAY PART WISE)#

```python
uber_city = uber.groupby(["Status", "Day_part"]).count()['Request id'].unstack().fillna(0)
```

```python
uber_city.plot(kind='bar', stacked=True, align='center', title='Uber status from city to airport')
```

```python
uber_airport = uber.groupby(["Status", "Day_part"]).count()['Request id'].unstack().fillna(0)
uber_airport.plot(kind='bar', stacked=True, align='center', title='Uber status from airport to city')
```

```python
#Plotting pie chart for status
uber.Status.value_counts().plot(kind="pie",startangle = 90,radius=1.0,autopct="%0.0f%%")
plt.title("Request Status")
plt.tight_layout()
plt.axis("equal")
plt.show()
```

```python
#Identifying Supply and Demand GAP from City to Airport
#DAY_PART(PART OF THE DAY)
```

```python
#Status-1 --> No Cars Available

uber_gap =  uber[uber['Status'] == 'No Cars Available']

uber_gap_citytoairport = uber_gap.loc[uber_gap['Pickup point'] == 'City']

uber_gap_citytoairport = uber_gap_citytoairport.groupby(['Day_part', 'Status'])['Status'].count().unstack('Status').fillna(0)
uber_gap_citytoairport.plot(kind='bar', stacked=True, title = 'Supply and Demand Gap from City to Airport')
```

```python
#Status-2 --> Cancelled

uber_gap =  uber[uber['Status'] == 'Cancelled']

uber_gap_citytoairport = uber_gap.loc[uber_gap['Pickup point'] == 'City']

uber_gap_citytoairport = uber_gap_citytoairport.groupby(['Day_part', 'Status'])['Status'].count().unstack('Status').fillna(0)
uber_gap_citytoairport.plot(kind='bar', stacked=True, title = 'Supply and Demand Gap from City to Airport')
```
